```python
def calculate_total_price(item_price, items_purchased):
  return item_price * items_purchased

#main

def main():
  item_price = float(input("Enter the price of the item: "))
  items_purchased = int(input("Enter the number of items purchased: "))
  total_price = calculate_total_price(item_price, items_purchased)
  print(f"The total price is: ${total_price:.2f}")

if __name__ == "__main__":
  main()
```
